```python
import pandas as pd
pd.options.mode.copy_on_write = True
```

```python
data_set = 'blood'
df = pd.read_csv(f'{data_set}.csv')
df
```

```python
filtered_data_with_100_epoch = df[df['epoch'] == 100]

filtered_data_with_100_epoch
```

```python
# organamnist
organlabel_dict = {
    'class0': 'bladder',
    'class1': 'femur-left',
    'class2': 'femur-right',
    'class3': 'heart',
    'class4': 'kidney-left',
    'class5': 'kidney-right',
    'class6': 'liver',
    'class7': 'lung-left',
    'class8': 'lung-right',
    'class9': 'pancreas',
    'class10': 'spleen'
}

# pneumoniamnist
pneumonialabel_dict = {
    'class0': "normal",
    'class1': "pneumonia",
}

# bloodmnist
bloodlabel_dict = {
    'class0': 'basophil',
    'class1': 'eosinophil',
    'class2': 'erythroblast',
    'class3': 'immature granulocytes',
    'class4': 'lymphocyte',
    'class5': 'monocyte',
    'class6': 'neutrophil',
    'class7': 'platelet'
}
```

```python
def select_label_dict(data_set_name):
    if data_set_name == 'organa':
        return organlabel_dict
    elif data_set_name == 'pneumonia':
        return pneumonialabel_dict
    elif data_set_name == 'blood':
        return bloodlabel_dict
    else:
        return {}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

embedding_zero_data = filtered_data_with_100_epoch[filtered_data_with_100_epoch['embedding'] == 0]
line_data = filtered_data_with_100_epoch[filtered_data_with_100_epoch['embedding'] != 0]

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))

sns.lineplot(x='embedding', y='unconditional_fid', hue='gan', style='gan', markers=True, dashes=False, data=line_data)

for gan_type in embedding_zero_data['gan'].unique():
    gan_data = embedding_zero_data[embedding_zero_data['gan'] == gan_type]
    marker_style = 'o' if gan_type == 'vanilla' else 'X'  
    plt.scatter('embedding', 'unconditional_fid', data=gan_data, label=f'{gan_type} (One-Hot)', 
                marker=marker_style, s=100, zorder=5)


plt.title(f'{data_set} Unconditional FID vs. Embedding Size for Different GANs (Distinct One-Hot)')
plt.xlabel('Embedding Size')
plt.ylabel('Unconditional FID')
plt.legend(title='GAN Type and Condition')

plt.savefig(f'Figures/{data_set}_unconditional_embedding.png')


plt.show()
plt.close()
```

```python
import numpy as np
conditional_fid_columns = [col for col in filtered_data_with_100_epoch.columns if 'class' in col]
long_form_data = pd.melt(filtered_data_with_100_epoch, id_vars=['gan', 'embedding'], value_vars=conditional_fid_columns,
                         var_name='class', value_name='conditional_fid')

long_form_data['conditional_fid'] = pd.to_numeric(long_form_data['conditional_fid'], errors='coerce')

vanilla_data = long_form_data[long_form_data['gan'] == 'vanilla']
wasserstein_data = long_form_data[long_form_data['gan'] == 'Wasserstein']

y_min = np.min(long_form_data['conditional_fid'])*0.9
y_max = np.max(long_form_data['conditional_fid'])*1.1

```

```python
label_dict = select_label_dict(data_set)
vanilla_data.loc[:, 'class_name'] = vanilla_data['class'].map(label_dict)
wasserstein_data.loc[:, 'class_name'] = wasserstein_data['class'].map(label_dict)
```

```python
plt.figure(figsize=(14, 16))

class_markers = ['o', 's', '^', 'P', '*', 'X', 'D', 'v','+','<','>']

plt.subplot(2, 1, 1)
sns.lineplot(x='embedding', y='conditional_fid', hue='class_name', data=vanilla_data[vanilla_data['embedding'] != 0], markers=True, dashes=False)

for i, class_label in enumerate(vanilla_data['class'].unique()):
    class_name = label_dict[class_label]
    class_data = vanilla_data[(vanilla_data['embedding'] == 0) & (vanilla_data['class'] == class_label)]
    plt.scatter('embedding', 'conditional_fid', data=class_data, label=f'{class_name} (One-Hot)', 
                marker=class_markers[i], s=100, zorder=5)

plt.title(f'{data_set} Conditional FID vs. Embedding Size for Vanilla GAN')
plt.xlabel('Embedding Size')
plt.ylabel('Conditional FID')
plt.ylim(y_min, y_max)
plt.legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.subplot(2, 1, 2)
sns.lineplot(x='embedding', y='conditional_fid', hue='class_name', data=wasserstein_data[wasserstein_data['embedding'] != 0], markers=True, dashes=False)

for i, class_label in enumerate(wasserstein_data['class'].unique()):
    class_name = label_dict[class_label]
    class_data = wasserstein_data[(wasserstein_data['embedding'] == 0) & (wasserstein_data['class'] == class_label)]
    plt.scatter('embedding', 'conditional_fid', data=class_data, label=f'{class_name} (One-Hot)', 
                marker=class_markers[i], s=100, zorder=5)

plt.title(f'{data_set} Conditional FID vs. Embedding Size for Wasserstein GAN')
plt.xlabel('Embedding Size')
plt.ylabel('Conditional FID')
plt.ylim(y_min, y_max)
plt.legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

plt.savefig(f'Figures/{data_set}_conditional_embedding.png')

plt.show()
plt.close()
```

```python
conditional_fid_columns = [col for col in filtered_data_with_100_epoch.columns if 'class' in col]
unconditional_fid_data = filtered_data_with_100_epoch[['embedding', 'unconditional_fid','gan'] + conditional_fid_columns]

melted_data = pd.melt(unconditional_fid_data, id_vars=['embedding', 'unconditional_fid','gan'], value_vars=conditional_fid_columns,
                      var_name='class', value_name='class_fid')

melted_data['embedding'] = melted_data['embedding'].replace(0, 'one-hot')
melted_data['embedding'] = melted_data['embedding'].astype(str)
melted_data['class'] = melted_data['class'].astype(str)
melted_data.loc[:, 'class_name'] = melted_data['class'].map(label_dict)
melted_data['gan_embedding'] = melted_data.apply(lambda x: f"{x['gan']} (Embedding {x['embedding']})", axis=1)

melted_data
```

```python
gan_embedding_order = sorted(melted_data['gan_embedding'].unique(), key=lambda x: ('one-hot' in x, x))
plt.figure(figsize=(14, 8))

class_fid_min = melted_data['class_fid'].min()
class_fid_max = melted_data['class_fid'].max()
sns.barplot(x='class_name', y='class_fid', hue='gan_embedding', data=melted_data,hue_order=gan_embedding_order)

plt.ylim(class_fid_min * 0.9, class_fid_max * 1.1)
plt.title(f'{data_set} Unconditional FID vs Class for Different Embedding Sizes')
plt.xlabel('Class')
plt.ylabel('Class FID')
plt.legend(title='Embedding Size')

plt.xticks(rotation=25, fontsize='small')
plt.tight_layout()

plt.savefig(f'Figures/{data_set}_conditional_class.png')
plt.show()
```
